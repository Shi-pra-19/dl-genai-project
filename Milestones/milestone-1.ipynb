```python
import pandas as pd

df = pd.read_csv("train.csv")

label_cols = ['anger', 'fear', 'joy', 'sadness', 'surprise']


emotion_counts = df[label_cols].sum()

most_common = emotion_counts.idxmax()
count_most_common = emotion_counts.max()

print("Most common emotion:", most_common)
print("Count:", count_most_common)
```

```python
df.head()
```

```python
df['label_count'] = df[label_cols].sum(axis=1)

exactly_two = (df['label_count'] == 2).sum()
print("Rows with exactly 2 labels:", exactly_two)
```

```python
both = df[(df['joy'] == 1) & (df['sadness'] == 1)].shape[0]
print("Rows with Joy and Sadness together:", both)
```

```python
surprise_pct = (df['surprise'] == 1).mean() * 100
print("Percentage Surprise:", round(surprise_pct, 2))
```

```python
max_diff = emotion_counts.max() - emotion_counts.min()
print("Max difference:", max_diff)
```

```python
df['word_count'] = df['text'].str.split().apply(len)
median_words = df['word_count'].median()
print("Median word count:", median_words)
```

```python
corr_val = df['anger'].corr(df['fear'])
print("Correlation:", round(corr_val, 2))
```

```python
import string
import re

def clean_text(t):
    t = t.lower()  # lowercase
    t = t.translate(str.maketrans("", "", string.punctuation))  # remove punctuation
    return t

df['clean_text'] = df['text'].apply(clean_text)
```

```python
orig_chars = df['text'].str.len().sum()
clean_chars = df['clean_text'].str.len().sum()

pct_reduction = (orig_chars - clean_chars) / orig_chars * 100
print("Character reduction %:", round(pct_reduction, 2))
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop = set(stopwords.words('english'))
```

```python
words = " ".join(df['clean_text']).split()
unique_words = set(words)

stop_unique = unique_words.intersection(stop)

pct_stop = len(stop_unique) / len(unique_words) * 100
print("Percentage unique stopwords:", round(pct_stop, 2))
```

```python
from collections import Counter

filtered_words = [w for w in words if w not in stop]

freq = Counter(filtered_words)
fifth_word = freq.most_common(5)[4][0]

print("5th most frequent word:", fifth_word)
```
