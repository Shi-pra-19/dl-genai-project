#DATA CLEANING

## Overview

This notebook performs data cleaning on a text column. Duplicate rows based on the text column are removed to ensure each text entry is unique. Then, a set of unwanted target strings ("& # x 26; nbsp;", "#name?", "##url##", ">", "<", etc) are removed from the text column in both datasets. After cleaning, we check for empty text rows in both the training and test sets, identifying a few entries with no text. Finally, it saves the cleaned datasets to new CSV files: train_clean.csv and test_clean.csv.

Results:

- Training data shape: (4986, 8)

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/test.csv')
```

**Remove duplicates, keeping only set of text**

```python
df = df.drop_duplicates(subset='text', keep='first')
```

```python
df.shape
```

**Remove unnecessary words**

```python
targets = [
    "& # x 26; nbsp;",
    "#name?",
    "##url##",
    "&gt;",
    "&lt;"
]
for t in targets:
    df["text"] = df["text"].str.replace(t, "", regex=False)
    test["text"] = test["text"].str.replace(t, "", regex=False)
```

```python
df.shape
```

```python
empty_rows = df[df["text"]== ""]

print(empty_rows)
```

```python
empty_rows = test[test["text"]== ""]

print(empty_rows)
```

**Save**

```python
df.to_csv('train_clean.csv', index=False)
test.to_csv('test_clean.csv', index=False)
```
